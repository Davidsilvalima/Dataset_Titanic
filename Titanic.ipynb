```python
def plot_bar(variavel,
             data,
             figsize = (10, 5),
             bins = 30):
  values = data[variavel].value_counts()
  x = values.index
  plt.figure(figsize = figsize)
  plt.bar(x, values)
  plt.ylabel('Frequência')
  plt.xticks(x, values.index.values)
  plt.title(variavel)
  plt.grid()
  plt.show()
```

```python
print(plot_bar)
```
